### Hyperparameter tuning


In my  case study I finally evaluate the follwing three parameters .
1. What is the right size of the hidden layer or layers?_
2. What is the right activation function for this problem?_
3. What is the optimal learning rate to find the optimal model for this problem?

Here is that `print results` function that is just going to help us explore the results of this GridSearchCV.

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

 I have called the `MLPClassifier` object and stored it as `mlp`. I  hardcoded the hyperparamter values in here by entering them in the parentheses but I tested them using `GridSearchCV` 
I defined the hyperparameter dectionary. So the keys in this dictionary align with the name of the hyperparameters that would be passed into `MLPClassifier` and then we just need to define the list of settings we want to test.

So for hidden layer size, I only be explored models with ONE hidden layer. So we'll look at one layer with 10 nodes, one layer with 50 nodes, and one layer with 100 nodes. so the second entry in this tuple would represent the number of layers.

Next, I  explored the three activation functions`relu` (this is the default), `tanh`, and `logistic` which is the same as the sigmoid .

Lastly, I looked at the learning rate. Now here are a few options given in `sklearn` 
- _`constant` is just keeping a constant `learning rate` given by a different hyperparameter_
- _`invscaling` gradually decreases the learning rate at each step, so this will allow it to take large jumps at first and then slowly decrease it hopefully as it gets closer and closer to the optimal model_
- _`adaptive` keeps the learning rate constant as long as training loss keeps decreasing, if it STOPS decreasing then it will decrease the learning rate to take smaller steps_

So next I called `GridSearchCV`, pass in my model object (`mlp`), the hyperparameter dictionary, and tell it we want to do 5-fold Cross Validation._

I just called `.fit()` with my training features and labels and that will prompt it to grab the first hyperparameter combination, so that would be one hidden layer with 10 nodes, using the `relu` activation function, and a constant learning rate. And it will use that setting and run Cross Validation. It will do this for each hyperparameter combination._

_So lets look at the results._

```python
mlp = MLPClassifier()
parameters = {
    'hidden_layer_sizes': [(10,), (50,), (100,)],
    'activation': ['relu', 'tanh', 'logistic'],
    'learning_rate': ['constant', 'invscaling', 'adaptive']
}
cv = GridSearchCV(mlp, parameters, cv=5)
cv.fit(tr_features, tr_labels)

print_results(cv)
```
